```python
# BMI Category:
df["BMI Category"] = np.where(df['Body mass index'] < 25, "Underweight",
                              np.where(df['Body mass index'] < 30, "Normal Weight",
                                       np.where(df['Body mass index'] < 35, "Over Weight", "Obese")))
```

```python
print(df["BMI Category"])
```

```python
df["BMI Category"].head(10)
```

```python
# Create an Age Group Column
df["Age Group"] = np.where(df['Age'] < 35, "Young",
                           np.where(df['Age'] < 40, "Adult",
                                    np.where(df['Age'] < 45, "Middle Aged", "Senior")))
```

```python
print(df["Age Group"])
```

```python
df["Age Group"].value_counts()
```

```python
df.columns
```

```python
df.shape
```

```python
# Create  Season group
df["Seasons Name"] = np.where(df['Seasons'] == 1, "Spring",
                              np.where(df["Seasons"] == 2 , "Summer",
                                       np.where(df['Seasons'] ==3, "Autumn", "Winter")))
```

```python
print(df["Seasons Name"])
```

```python
df["Seasons Name"].value_counts()
```

```python
df['education_level']=np.where(df['Education']==1,'High School',
                               np.where(df['Education']==2,'Graduate',
                                        np.where(df['Education']==3,'Post Graduate','Doctorate')))
```

```python
print(df['education_level'].value_counts())
```

```python
df.shape
```

```python
# create a new column by using

df['pet_column']=np.where(df['Pet']==0,"no pets",
                          np.where(df['Pet']<2,"having 1 pet","having 2 pets"))
print(df['pet_column'].value_counts())
```
