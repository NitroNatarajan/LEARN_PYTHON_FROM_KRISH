```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data_set/diverseTalent.csv")

# Data visualization 

# how many duplicate rows in the dat set
df.duplicated().sum()

# Display of the duplicated data 
duplicated = df[df.duplicated()]

# display the distribution of the gender column 
gender_count = df["Gender"].value_counts()

# What are the unique values in the edu back column 
edu_unique = df["Educational_Background"].unique()
edu_unique

# Visualize the distribution of the experience_level. countplot 
plt.figure(figsize = (10, 5))
sns.countplot(data = df, x = "Experience_Level")
plt.title("Distribution of Experience Level")
plt.xlabel("Experience_Level")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df , x="Application_Status")
plt.title("Distribution of Application Status")
plt.xlabel("Application Status")
plt.ylabel("Frequency")
plt.show()
```

```python
# what is the correlation between numerical columns
numeric_columns = df.select_dtypes(include=["int64","float"])
# numeric_columns.isnull().sum()
corr_matrix = numeric_columns.corr()
sns.heatmap(corr_matrix)
plt.show()
```

```python
#  Analysis using where()

df["High_Salary"] = np.where(df["Salary_Expectation"]>100000, "YES" , "NO")
df["High_Salary"].value_counts()
```

```python
df["Experienced"] = np.where(df["Experience_Level"] == 0 , "Fresher", 
                             np.where(df["Experience_Level"] < 5 , "Senior" , "Super_Senior"))
df["Experienced"].value_counts()
df["Interview_Score"]
```

```python
df["Good_Score"] = np.where(df["Interview_Score"] > 80, "Yes" ,"No" )
df["Good_Score"].value_counts()
```

```python
df["Certified"] = np.where(df["Certifications_Count"] >=2 , "Yes" , "No")

# Q25: Create a column 'Technical_Background' where the value is 

#'Yes' if 'Educational_Background' is 'Computer Science' or 'Information Technology'. 

#Otherwise, 'No'.
df["Educational_Background"]
df["Technical_Background"] = np.where((df["Educational_Background"] == "Computer Science") | (df["Educational_Background"] == "Engineering"), "Yes" , "No" )
df["Technical_Background"].value_counts()
df["Age", "Region"]
```
